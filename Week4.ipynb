# 2. Weekly graph question

Below are a histogram and boxplot representation of the same data. A pharmacy is keeping a record of the prices of the drugs that it sells, and an administrator wants to know how much the more expensive drugs tend to cost, in the context of the other prices.

Please write a short explanation of the pros and cons of these two representations. Which would you choose? How would you modify the formatting, if at all, to make it more visually interesting, clear, or informative?

```python
import numpy as np
import pandas as pd
np.random.seed(0)
num_data = 100
data = np.exp(np.random.uniform(size = num_data) * 4)
df = pd.DataFrame(data.T, columns = ["data"])
```

```python
print("The 75th percentile is:", df.quantile(q = 0.75))
df.plot.hist()
```

```python
df.plot.box()
```

The histogram and boxplot are two different ways of summarizing and visualizing the distribution of the data. Both have their pros and cons when it comes to providing insights into the prices of drugs.

Histogram:
- Pros:
  - Shows the distribution of the data (i.e., how the data points are spread across different ranges).
  - Helps in understanding the frequency of prices in different intervals.
  - Makes it easy to see the shape of the data, including whether it's skewed, bimodal, or has outliers.
  
- Cons:
  - Can be sensitive to bin sizes, and different choices of bins might affect how the data is perceived.
  - Exact values of specific data points are not clearly shown (only the frequency in bins).
  - Not as effective in highlighting summary statistics like median or percentiles directly.

Boxplot:
- Pros:
  - Provides a summary of the data with the median, quartiles, and outliers clearly visible.
  - Highlights the central tendency (median) and spread of the data, along with any extreme values (outliers).
  - It’s particularly useful for quickly identifying skewness or asymmetry in the data distribution.
  
- Cons:
  - Doesn’t show the full distribution of data (no information on frequency or shape).
  - The appearance of a boxplot may be confusing if the data contains many outliers or is very skewed.

Which One to Choose?
Given that the administrator wants to know how much the more expensive drugs tend to cost, the boxplot might be more useful for this context. It clearly shows the 75th percentile, which directly answers the question about the higher end of the prices. The boxplot also reveals any extreme outliers, which may be the particularly expensive drugs in this case.

Modifications for Better Clarity and Aesthetics:
1. Histogram:
   - Adjust the bin width** for better granularity and to avoid over-binning or under-binning.
   - Add labels for better clarity, such as axis labels and a title.
   - Consider using a density plot overlay to give a smoother representation of the distribution.
   - Use color to make the histogram more visually appealing.

2.Boxplot:
   - Add labels or a title to clarify that the boxplot represents drug prices.
   - Show outliers clearly (if there are any), perhaps by adjusting their marker size or color for emphasis.
   - Make sure the axes are labeled for clarity (e.g., “Price of Drugs”).

Conclusion:
Both the histogram and the boxplot have valuable insights to offer, but for the administrator's specific question regarding more expensive drugs, the boxplot would be the preferred option. To improve the visuals, I would ensure both plots have clear labeling, proper axis scales, and consider enhancing the histogram with a density curve for a smoother presentation.

```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
df = sns.load_dataset("titanic")
print(df.head())
df['age'].plot.hist(bins=20, edgecolor='black', alpha=0.7)
plt.title("Distribution of Passenger Ages")
plt.xlabel("Age")
plt.ylabel("Count")
plt.show()
sns.kdeplot(df['fare'], shade=True)
plt.title("Fare Distribution (KDE Plot)")
plt.xlabel("Fare ($)")
plt.show()
sns.violinplot(x=df["fare"])
plt.title("Violin Plot of Fare Prices")
plt.show()
plt.figure(figsize=(8, 6))
sns.swarmplot(x=df["class"], y=df["age"])
plt.title("Swarm Plot of Age by Passenger Class")
plt.xlabel("Passenger Class")
plt.ylabel("Age")
plt.show()
sns.histplot(data=df, x="age", hue="survived", bins=20, kde=True)
plt.title("Age Distribution by Survival")
plt.xlabel("Age")
plt.ylabel("Count")
plt.legend(["Did not survive", "Survived"])
plt.show()
sns.countplot(x=df['class'])
plt.title("Passenger Count by Class")
plt.xlabel("Class")
plt.ylabel("Count")
plt.show()
```

```python
print("Age - Mean:", df['age'].mean())
print("Age - Median:", df['age'].median())
print("Age - Mode:", df['age'].mode()[0])
print("Age - Min:", df['age'].min())
print("Age - Max:", df['age'].max())

print("Fare - Mean:", df['fare'].mean())
print("Fare - Median:", df['fare'].median())
print("Fare - Mode:", df['fare'].mode()[0])
print("Fare - Min:", df['fare'].min())
print("Fare - Max:", df['fare'].max())
```

# 4. Storytelling With Data graph

Reproduce any graph of your choice in p. 52-68 of the Storytelling With Data book as best you can.  (The second half of chapter two).  You do not have to get the exact data values right, just the overall look and feel.

```python
import matplotlib.pyplot as plt
import numpy as np
categories = ["Start", "Sales", "Marketing", "R&D", "Operations", "Net Profit"]
values = [100, 30, -20, -10, -15, 85]  # Changes in values
start_values = np.zeros(len(values))
for i in range(1, len(values)):
    start_values[i] = start_values[i - 1] + values[i - 1]
colors = ["blue" if i == 0 else "green" if val > 0 else "red" for i, val in enumerate(values)]
fig, ax = plt.subplots(figsize=(8, 5))
ax.bar(categories, values, bottom=start_values, color=colors, edgecolor='black')
for i, (val, start) in enumerate(zip(values, start_values)):
    ax.text(i, start + val/2, f"{val:+}", ha="center", va="center", fontsize=12, color="white" if abs(val) > 10 else "black")
plt.title("Waterfall Chart - Financial Breakdown")
plt.ylabel("Value in $")
plt.grid(axis="y", linestyle="--", alpha=0.7)
plt.show()
```
